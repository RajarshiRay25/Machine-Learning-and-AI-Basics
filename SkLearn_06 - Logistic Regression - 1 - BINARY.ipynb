```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Access the csv
new_data = pd.read_csv("insurance_data.csv")
new_data
```

```python
# Graph of data values -- sigmoid -- logistic regression

plt.xlabel("Age")
plt.ylabel("Insurance Bought")
plt.scatter(new_data.age,new_data.bought_insurance,marker="+",color="red")
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(new_data[["age"]],new_data.bought_insurance,train_size=0.8)
```

```python
X_test
```

```python
# Import Logistic Regression

from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()  # Logistic Regression Class
```

```python
# Train the model 
log_reg.fit(X_train,Y_train)
```

```python
# Predict the values
log_pred = log_reg.predict(X_test)
log_pred
```

```python
# check the score of model
log_reg.score(X_test,Y_test)
```

```python
# Get a No vs Yes probability

log_reg.predict_proba(X_test)
```
