# Mathematical Ops

```python
import numpy as np
```

```python
arr_01 = np.arange(1,10).reshape(3,3)
arr_02 = np.arange(1,10).reshape(3,3)
print(arr_01)
print(arr_02)
```

```python
# Addition
arr_01+arr_02
```

```python
arr_add = np.add(arr_01,arr_02)
print(arr_add)
```

```python
# Subtraction
arr_01-arr_02
```

```python
arr_sub = np.subtract(arr_01,arr_02)
print(arr_sub)
```

```python
# Multiply
arr_01*arr_02
```

```python
arr_mul = np.multiply(arr_01,arr_02)
print(arr_mul)
```

```python
# Division
arr_01/arr_02
```

```python
arr_div = np.divide(arr_01,arr_02)
print(arr_div)
```

```python
# Matrix Multiplication
arr_01@arr_02
```

```python
arr_mat_mul = arr_01.dot(arr_02)
print(arr_mat_mul)
```

# Max,Min and Numerical OPS

```python
# max element from array
arr_test = np.arange(1,10).reshape(3,3)
arr_test.max()
```

```python
# max index 
arr_test.argmax()
```

```python
# max from col or row
# axis=0 -> col
# axis=1 -> row
arr_test = np.arange(1,10).reshape(3,3)
arr_test
print(arr_test)
print(arr_test.max(axis=0))
print(arr_test.max(axis=1))
```

```python
# min from col or row
# axis=0 -> col
# axis=1 -> row
arr_test = np.arange(1,10).reshape(3,3)
arr_test
print(arr_test)
print(arr_test.min(axis=0))
print(arr_test.min(axis=1))
```
