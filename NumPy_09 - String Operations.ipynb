```python
import numpy as np
```

```python
myStr = "Hello Welcome to Machine Learning"
myStr2 = "Starting with NumPy \n Moving to Pandas next."
```

```python
np.char.add(myStr,myStr2)   # Concatenate 
```

```python
np.char.lower(myStr)   # Lower Case
```

```python
np.char.upper(myStr2)  # Upper Case
```

```python
np.char.center(myStr,90,fillchar="-")    # Center aligns
```

```python
np.char.split(myStr2)  # Split words
```

```python
np.char.splitlines(myStr2) 
```

```python
np.char.splitlines(myStr2)   # Split according to lines
```

```python
np.char.equal(myStr,myStr2)  # Checks equality
```
