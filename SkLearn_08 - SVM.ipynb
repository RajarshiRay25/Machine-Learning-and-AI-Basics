```python
# target means numerical rep of each flower
iris.target
```

```python
# Create a dataset frame
iris_df = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df.head()
```

```python
# Create a target dataframe
iris_df["target"] = iris.target
iris_df.head()
```

```python
iris_df[iris_df.target==1].head()
```

```python
# set a new column

iris_df["flower_name"]=iris_df.target.apply(lambda x:iris.target_names[x])
```

```python
iris_df.head()
```

```python
# Searching for all the flowers

iris_setosa = iris_df[iris_df.target==0]   # setosa flower start from 0
iris_versi = iris_df[iris_df.target==1]   # versicolor start frpm 50
iris_virgin = iris_df[iris_df.target==2]   # virginica start frpm 100
```

```python
# we will graphically compare sepal length and width

import matplotlib.pyplot as plt
```

```python
plt.xlabel("Length of Sepal")
plt.ylabel("Width of Sepal")
# setosa compare
plt.scatter(iris_setosa["sepal length (cm)"],iris_setosa["sepal width (cm)"],color="red",marker="*")

# versicolor compare

plt.scatter(iris_versi["sepal length (cm)"],iris_versi["sepal width (cm)"],color="blue",marker="+")
```

```python
plt.xlabel("Length of Sepal")
plt.ylabel("Width of Sepal")
# setosa compare
plt.scatter(iris_setosa["sepal length (cm)"],iris_setosa["sepal width (cm)"],color="red",marker="*")

# virginia compare

plt.scatter(iris_virgin["sepal length (cm)"],iris_versi["sepal width (cm)"],color="blue",marker="+")
```

```python
# Train_Test
from sklearn.model_selection import train_test_split
```

```python
iris_df
```

```python
# create train data - only values
iris_X = iris_df.drop(["target","flower_name"],axis="columns")
iris_Y = iris_df.target
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(iris_X,iris_Y,test_size=0.2)
```

```python
# create SVC -- SVM
from sklearn.svm import SVC
# X - petal and sepal value
# Y - target
```

```python
iris_model = SVC()
```

```python
iris_model.fit(X_train,Y_train)
```

```python
iris_model.score(X_test,Y_test)
```

```python
iris_model.predict([[4.1,2.9,5.0,3.9]])
```
