```python
import numpy as np
import random
```

```python
np.random.random(1)   # Returns 1 random value
```

```python
np.random.random((4,4))  # Returns 2D array
```

```python
np.random.randint(1,9,(4,4))  # (start,end,matrix type)
```

```python
np.random.randint(1,9,(2,4,4)) # (start,end,(n number,matrix type))
```

```python
np.random.rand(1,2,3,4)  # returns val between 0 and 1
```

```python
list = [1,2,3,4,5,6]
np.random.choice(list)
```

```python
np.random.permutation(list)
```
