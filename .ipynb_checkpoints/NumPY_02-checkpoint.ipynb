```python
import numpy as np
```

```python
new_Matrix01 = np.array([[1,1,1],  # Ones matrix
          [1,1,1],
          [1,1,1]])
print(new_Matrix01)
```

```python
matrix_Ones = np.ones(5)  # Ones matrix rows
print(matrix_Ones)
```

```python
matrix_Ones.dtype  # Data type
```

```python
newOnes = np.ones((3,3))   # Ones matrix row vs col
print(newOnes)
```

```python
assignMat = np.ones((3,3),dtype= int)
print(assignMat)
```

```python
matrix_Zeros = np.zeros((4,4))  # Zeros matrix
print(matrix_Zeros)
```

```python
matrix_Zeros = np.zeros((4,4),dtype= bool)  # BOOLEAN -> 1 - True : 0 - False
print(matrix_Zeros)
```

```python
empty_Matrix = np.empty((3,3),dtype = float)  # Create empty matrix random
print(empty_Matrix)
```
