```python
import numpy as np
```

```python
# arange function -> range print array
# start,end,steps

arr_1D = np.arange(0,15,2)
print(arr_1D)
```

```python
# reshape function -> Reshape dim of array  
# 3D,row,columns

arr_res = arr_1D.reshape(2,4)
print(arr_res)
```

```python
arr_res = arr_1D.reshape(2,2,2)
print(arr_res)
```

```python
# ravel function -> conv nD array -> 1D array
# flatten function
arr_conv = arr_res.ravel()
print(arr_conv)
```

```python
# transpose matrix
# row -> col -> row

arr_res = arr_1D.reshape(2,4)
print(arr_res)

arr_res.transpose()
# or
arr_res.T
```
