## Series & DataFrame

```python
# Series and DataFrame

list_ser = [1,2,3,4]
dict_df = {
    "Name" : ["AR","BR","CR","DR"],
    "Age"  : [20,21,22,23],
    "City" : ["AX","BX","CX","DX"]
}
```

```python
series = pd.Series(list_ser,index=["First","Second","Third","Fourth"])
series      # Series with index
```

```python
dataframe_new = pd.DataFrame(dict_df)     # dataframe with index
dataframe_new.index = ["First","Second","Third","Fourth"]
dataframe_new.head()
```

## LOC & ILOC

```python
read_file.head()
```

```python
read_file.loc["First" , "Type"]    # loc used to access table based on their actual row and col name
```

```python
# to change a certain part use through loc or iloc
# changing Ray's Type to SDE
read_file.loc["Second","Type"] = "SDE"
read_file.head()
```

```python
read_file.drop("Second",axis=0)    # remove row second
```

```python
read_file.drop("Name",axis=1)     # remove Name column
```

```python
read_file.head()
```

```python
read_file.iloc[0,1]      # indexed based loc
```

```python
read_file.iloc[2,2] = "App"    # changing Sean EH to App
read_file.head()
```
