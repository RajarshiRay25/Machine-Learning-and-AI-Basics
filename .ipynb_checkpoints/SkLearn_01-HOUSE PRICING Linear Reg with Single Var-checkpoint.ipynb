```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# Import house data csv to test
house_data_file = pd.read_csv("house_data_csv.csv")
house_data_file
```

```python
# We will predict house price

plt.xlabel("Area Of Plot")
plt.ylabel("Price Of House")
plt.scatter(house_data_file["area"],house_data_file["price"],color="red",marker="+")
```

## Y=mX+c

### Y -> Price we want to see
### X -> Factors used
### m -> coeff
### c -> intercept

```python
# Lets to regression

reg = linear_model.LinearRegression()

# Fit the model --> fit(array of factors in 2D , what factor we want to predict)

reg.fit(house_data_file[["area"]],house_data_file.price)
```

```python
# predict values desired in -> predict([[factor]])

reg.predict([[3400]])
```

```python
reg.coef_  # print of coeff
```

```python
reg.intercept_  # print intercept
```

```python
# We will predict house price

plt.xlabel("Area Of Plot")
plt.ylabel("Price Of House")
plt.scatter(house_data_file["area"],house_data_file["price"],color="red",marker="+")
plt.plot(house_data_file["area"],reg.predict(house_data_file[["area"]]))   # see the prediction graph
```

### Lets Test  a CSV Data
