```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

## Find income for year 2020

```python
can_data = pd.read_csv("canada_per_capita_income.csv")
can_data
```

```python
plt.xlabel("Year")
plt.ylabel("income")
plt.scatter(can_data.year,can_data.income,color="red",marker="+")
```

```python
test_reg = linear_model.LinearRegression()
test_reg.fit(can_data[["year"]],can_data.income)
```

```python
test_reg.predict([[2020]])
```
