```python
my_list = [1,2,3,4,5,"pandas"]
print(my_list)
```

```python
series_1 = pd.Series(my_list)   # Makes a table of index vs values
print(series_1)
```

```python
series_emp = pd.Series([])    # Empty Series
print(series_emp)
```

```python
series_1_ind = pd.Series(my_list,index=['a','b','c','d','e','f'])  # Index gives label
print(series_1_ind)
```

```python
# name gives Table name by user
series_1_ind = pd.Series(my_list,index=['a','b','c','d','e','f'],name="My First Table")
print(series_1_ind)
```

```python
# Dictionary 
series_dict = pd.Series({
    "name" : "Ray",
    "age" : 20,
    "lang" : "Python"
})

print(series_dict)
```

```python
# Scalar Values
series_scalar = pd.Series(4)
print(series_scalar)
```
